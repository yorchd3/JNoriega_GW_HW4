```python
# Dependencies
import pandas as pd
```

```python
# load CSV
coders = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
coders_pd = pd.read_csv(coders)
coders_pd.head()
```

```python
# Inspect all columns
coders_pd.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_coders_pd = coders_pd.iloc[:, [
    0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_coders_pd.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
reduced_coders_pd = reduced_coders_pd.replace({0: "No", 1: "Yes"})
reduced_coders_pd.head()
```

```python
# Calculate total number of respondents in survey
respondents = len(reduced_coders_pd)
respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_bootcamp = reduced_coders_pd.loc[reduced_coders_pd["AttendedBootcamp"] == "Yes"]
attended_bootcamp.head()
```

```python
# Calculate average age of attendees
average_age = attended_bootcamp["Age"].mean()
print(average_age)

# Calculate how many people attended a bootcamp
amount_attended = attended_bootcamp["AttendedBootcamp"].count()
print(amount_attended)
```

```python

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
total_gender = attended_bootcamp["Gender"].count()
male = attended_bootcamp["Gender"].value_counts()['male']
female = attended_bootcamp["Gender"].value_counts()['female']
non_binary = total_gender - male - female
#print(f" Total: {total_gender}\n Male: {male}\n Female: {female}\n Non_Binary: {non_binary}")

# Calculate percentage of respondents who attended a bootcamp
percent_attended = amount_attended / respondents * 100
#print(percent_attended)

# Calculate percentage of respondents belonging to each gender
male_percent = (male/total_gender) * 100
female_percent = (female/total_gender) * 100
non_binary_percent = (non_binary/total_gender) * 100
print(f" % Male: {male_percent}\n % Female: {female_percent}\n % Non_Binary: {non_binary}")

# Calculate percentage of attendees with a school degree
#Remove a value non/degre
attended_bootcamp_degre = attended_bootcamp.loc[attended_bootcamp["SchoolDegree"] != "some college credit, no degree"]
attended_bootcamp_degre

# Calculate total number that has degree
holds_degree = len(attended_bootcamp_degre)
holds_degree

percent_degree = holds_degree/amount_attended*100
print(percent_degree)

# Calculate average post-bootcamp salary
avg_salary = attended_bootcamp["BootcampPostSalary"].mean()
print(avg_salary)
```

```python
# Create a new table consolodating above calculations
bootcamp_summary = pd.DataFrame({"Total Surveyed": [respondents],
                                   "Total Bootcamp attendees": [amount_attended],
                                   "% attended bootcamp": [percent_attended],
                                   "Avg. Age": [average_age],
                                   "% Male": [male_percent],
                                   "% Female": [female_percent],
                                   "% Non Binary": [non_binary_percent],
                                   "Has a degree": [percent_degree],
                                   "Average Post Bootcamp Salary": [avg_salary]
                                   })
bootcamp_summary
```

```python
# Improve formatting before outputting spreadsheet
bootcamp_summary = bootcamp_summary[["Total Surveyed",
                                         "Total Bootcamp attendees",
                                         "% attended bootcamp",
                                         "Avg. Age",
                                         "Has a degree",
                                         "% Male",
                                         "% Female",
                                         "% Non Binary",
                                         "Average Post Bootcamp Salary"]]
bootcamp_summary = bootcamp_summary.round(2)

bootcamp_summary["% Male"] = bootcamp_summary["% Male"].map(
    "{0:,.2f}%".format)
bootcamp_summary["% Female"] = bootcamp_summary["% Female"].map(
    "{0:,.2f}%".format)
bootcamp_summary["% attended bootcamp"] = bootcamp_summary["% attended bootcamp"].map(
    "{0:,.2f}%".format)
bootcamp_summary["% Non Binary"] = bootcamp_summary["% Non Binary"].map(
    "{0:,.2f}%".format)
bootcamp_summary["Has a degree"] = bootcamp_summary["Has a degree"].map(
    "{0:,.2f}%".format)
bootcamp_summary["Average Post Bootcamp Salary"] = bootcamp_summary["Average Post Bootcamp Salary"].map(
    "${0:,.0f}".format)
bootcamp_summary

bootcamp_summary
```

```python
# Export to Excel
bootcamp_summary.to_excel("output/BootcampOutputPart1_jorge.xlsx", index=False)
```
